```python
#Column wise concatenation
train_data=train_data.merge(train_labels,on="customer_ID")
train_data.to_csv("Train_1Mil.csv")
```

```python
data_copy = train_data.copy()
data_copy["S_2"] = pd.to_datetime(data_copy["S_2"])
data_copy["YYYY-MM"] = data_copy["S_2"].dt.to_period('M')
data_copy["YYYY-MM"]
data_copy.to_csv("MonthYearFull.csv")
```

```python
#one-hot encoding and dropping variables
#copying dataframe
encoded_data=train_data.copy()

#one hot encoding D_63,D_64 features
one_enc=pd.get_dummies(encoded_data["D_63"])
encoded_data=encoded_data.join(one_enc)

one_enc2=pd.get_dummies(encoded_data["D_64"])
encoded_data=encoded_data.join(one_enc2)

#dropping variables
encoded_data = encoded_data.drop("D_63",axis=1)
encoded_data = encoded_data.drop("D_64",axis=1)
```

```python
encoded_data.to_csv("Train_1MilOnehot.csv")
```

```python
print(encoded_data.shape)
encoded_data.head(15)
```

```python
train_data1 = encoded_data[(encoded_data['S_2'] >= '2017-05') & (encoded_data['S_2'] <= '2018-01')]
test_data1 = encoded_data[(encoded_data['S_2'] >= '2017-03-01') & (encoded_data['S_2'] <= '2017-04')]
test_data2 = encoded_data[(encoded_data['S_2'] >= '2018-02') & (encoded_data['S_2'] <= '2018-03') ]
```

```python
print(train_data1.shape)
print(test_data1.shape)
print(test_data2.shape)
```

```python
train_data1.to_csv("Train_main_1Mil.csv")
test_data1.to_csv("Test1_1Mil1.csv")
test_data2.to_csv("Test2_1Mil1.csv")
```
